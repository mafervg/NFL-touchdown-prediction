```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
years = ["2018", "2019", "2020", "2021", "2022"] 

dfs = []

for year in years:
    file_path = os.path.join(directory, f"play_by_play_{year}.csv")
    if os.path.exists(file_path):
        df = pd.read_csv(file_path, low_memory=False)
        dfs.append(df)
    else:
        print(f"File not found for year {year}")

# Concatenate all DataFrames into one
df = pd.concat(dfs, ignore_index=True)
```

### Explore features

- What correlates with touchdowns?<p>
- Use yards and previous season performance to predict touchdowns?

```python
#Let's take some basic quarteback stats

qb_feats = ['season', 'passer_id', 'passer', 'pass', 'complete_pass', 'interception', 'sack', 'yards_gained', 'touchdown']

#Let's group by season and who the passwer was
groupby_feats = ['season', 'passer_id', 'passer']

# Group by and aggregate by sum
qb_df = (df.loc[:,qb_feats]
        .groupby(groupby_feats, as_index=False)
        .sum())

#random sample of 10 player-seasons
qb_df.sample(10)
```

```python
#Let's see what correlates with touchdowns
for y in ['yards_gained', 'complete_pass', 'pass', 'interception', 'sack']:
    sns.regplot(data=qb_df, x='touchdown', y=y)
    plt.title(f"touchdowns and {y}")
    plt.show()
```

```python
df2 = qb_df.copy()

#add 1 to season
df2['season'] = df2['season'].add(1)

new_qb_df = (qb_df.merge(df2, on=['season', 'passer_id', 'passer'], 
                        suffixes=('', '_prev'),
                        how='left'))
new_qb_df.sample(10)
```

```python
#Let's see what correlates with next season touchdowns
for y in ['touchdown_prev', 'yards_gained_prev', 'complete_pass_prev', 'pass_prev', 'interception_prev', 'sack_prev']:
    sns.regplot(data=new_qb_df, x='touchdown', y=y)
    plt.title(f"touchdowns and {y}")
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
```

```python
new_qb_df.sample(10)
```

```python
#Let's use the previous season performance as our feautures from our model

features = ['pass_prev', 'complete_pass_prev', 'interception_prev', 'sack_prev', 'yards_gained_prev', 'touchdown_prev']

#We are predicting "current" season touchdowns
target = 'touchdown'

# Let's make a subset of data that has no null values
model_data = new_qb_df.dropna(subset=features + [target])

# Split the data into training (2018-2021) and testing (2022)
train_data = model_data[model_data['season'].isin([2018, 2019, 2020, 2021])]
test_data = model_data[model_data['season'] == 2022]

# Initialize the linear regression model
model = LinearRegression()

# Fit the model on the training data
model.fit(train_data[features], train_data[target])

# Make predictions on the test data
preds = model.predict(test_data[features])

# Create a Pandas Series for predictions with the same index as the test_data
preds = pd.Series(preds, index=test_data.index)

# Add the predictions as a new column 'preds' to the test_data DataFrame
test_data['preds'] = preds
```

```python
# Let's run some basic statistics to examine the
# quality of the prediction

rmse = mean_squared_error(test_data['touchdown'], test_data['preds'])**0.5
r2 = pearsonr(test_data['touchdown'], test_data['preds'])[0]**2
print(f"rmse: {rmse}\nr2: {r2}")
```

```python
# Let's visualize the outputs 
sns.regplot(data=test_data, x='touchdown', y='preds')
plt.title('touchdown and predictions')
plt.show()
```
